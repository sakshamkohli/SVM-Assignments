Q1. What is the mathematical formula for a linear SVM?

yi*[wtx+b]>=1

Q2. What is the objective function of a linear SVM?

Maximizing the margin between classes and minimizing the classification error.

Q3. What is the kernel trick in SVM?

The kernel trick in Support Vector Machines (SVMs) is a technique used to handle nonlinear classification problems by implicitly mapping the input data into a higher-dimensional feature space without explicitly computing the transformed feature vectors. This allows SVMs to effectively classify data that is not linearly separable in the original input space.

Q4. What is the role of support vectors in SVM Explain with example

The role of support vectors is to mark the margins that are most distance apart from each other. The support vector of a class is the data point nearest to the datapoints of the other class.

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

*Hyperplane* : A hyperplane in SVM is used to distinctly classify data points that are linearly seperable. This plane segregates the data points into classes.
*Marginal plane* : Marginal planes are guided by support vectors that are the nearest data points to other classes. The hyperplane passes directly between these marginal planes.
*Soft margin* : Soft margin is a flexible margin that provides the possibility of some misclassification.
*Hard margin* : Hard margin is a stricter margin that provides no possibility of misclassification.

Q6. SVM Implementation through Iris dataset.
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
X = iris.data[:, :2]  # Using only two features for visualization
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=42, test_size=.3)
```

```python
svm_classifier = SVC(kernel='linear')
```

```python
svm_classifier.fit(X_train, y_train)
```

```python
# Predict labels for the testing set
y_pred = svm_classifier.predict(X_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                    np.arange(y_min, y_max, 0.1))
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary')
plt.show()
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 
              'gamma': [0.1, 0.01, 0.001], 
              'kernel': ['linear', 'rbf']}
```

```python
grid_search = GridSearchCV(estimator=svm_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
accuracy = best_model.score(X_test, y_test)
```

```python
print("Best Hyperparameters:", best_params)
print("Accuracy:", accuracy)
```
